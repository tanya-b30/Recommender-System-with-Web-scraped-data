```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
numbers=[]
# Lists to store the scraped data in
titles = []
genres = []
directors = []
actors = []
plots = []
for i in range(51,2000,50):
    numbers.append(str(i))
for num in numbers:
    url = "https://www.imdb.com/search/title/?genres=romance&start="+num+"&explore=title_type,genres&ref_=adv_nxt"
    response = get(url)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')
    print(type(movie_containers))
    print(len(movie_containers))
    
    # Extract data from individual movie container
    for container in movie_containers:
        if container.find('div', class_ = 'ratings-metascore') is not None:
            title = container.h3.a.text
            titles.append(title)
            genre = container.find('div',attrs={"class":"lister-item-content"}).find('p').find('span', attrs={"class":"genre"}).text
            genres.append(str.lstrip(genre))
            director = container.find('div',attrs={"class":"lister-item-content"}).find('p',attrs={"class":""}).find('a').text
            directors.append(director)
            actor = container.find('div',attrs={"class":"lister-item-content"}).find('p',attrs={"class":""}).find_all('a')
            actor.pop(0)
            for j in range(len(actor)): 
                actor[j]= (actor[j].get_text())
            actors.append(actor)
            plot =  container.find('div',attrs={"class":"lister-item-content"}).find_all('p',attrs={"class":"text-muted"})[1].text
            plots.append(str.lstrip(plot))

movies_df = pd.DataFrame({'Title': titles,
'Genre': genres,
'Director': directors,
'Actors': actors,
'Plot': plots
})
movies_df["Genre"]
print(movies_df.info())
movies_df.head(10)
```
